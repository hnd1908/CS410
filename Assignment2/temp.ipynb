```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import t

# Tham s
df = 20         # Bc t do (degrees of freedom)
t_stat = 2.5     # Gi tr t-stat quan st c
alpha = 0.05     # Mc  ngha

# To phn phi t
x = np.linspace(-4, 4, 500)
y = t.pdf(x, df)

# Tnh p-value (two-tailed)
p_value = 2 * (1 - t.cdf(abs(t_stat), df))

# V biu 
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f"Phn phi t (df={df})", color='blue')

# Vng p-value (two-tailed)
plt.fill_between(x, y, where=(x >= t_stat), color='red', alpha=0.5, label=f"p-value/2 = {p_value/2:.3f}")
plt.fill_between(x, y, where=(x <= -t_stat), color='red', alpha=0.5)
plt.axvline(x=t_stat, color='black', linestyle='--', label=f"t-stat = {t_stat:.1f}")
plt.axvline(x=-t_stat, color='black', linestyle='--')

# ng critical value (=0.05)
t_critical = t.ppf(1 - alpha/2, df)
plt.axvline(x=t_critical, color='green', linestyle=':', label=f"Critical value (=0.05) = {t_critical:.2f}")
plt.axvline(x=-t_critical, color='green', linestyle=':')

plt.title(f"Two-tailed t-test\np-value = {p_value:.3f} (t-stat = {t_stat:.1f}, df={df})")
plt.xlabel("Gi tr t")
plt.ylabel("Mt  xc sut")
plt.legend()
plt.grid()
plt.show()
```
