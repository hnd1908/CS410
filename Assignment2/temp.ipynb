```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import t

# Tham số
df = 20         # Bậc tự do (degrees of freedom)
t_stat = 2.5     # Giá trị t-stat quan sát được
alpha = 0.05     # Mức ý nghĩa

# Tạo phân phối t
x = np.linspace(-4, 4, 500)
y = t.pdf(x, df)

# Tính p-value (two-tailed)
p_value = 2 * (1 - t.cdf(abs(t_stat), df))

# Vẽ biểu đồ
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f"Phân phối t (df={df})", color='blue')

# Vùng p-value (two-tailed)
plt.fill_between(x, y, where=(x >= t_stat), color='red', alpha=0.5, label=f"p-value/2 = {p_value/2:.3f}")
plt.fill_between(x, y, where=(x <= -t_stat), color='red', alpha=0.5)
plt.axvline(x=t_stat, color='black', linestyle='--', label=f"t-stat = {t_stat:.1f}")
plt.axvline(x=-t_stat, color='black', linestyle='--')

# Đường critical value (α=0.05)
t_critical = t.ppf(1 - alpha/2, df)
plt.axvline(x=t_critical, color='green', linestyle=':', label=f"Critical value (α=0.05) = ±{t_critical:.2f}")
plt.axvline(x=-t_critical, color='green', linestyle=':')

plt.title(f"Two-tailed t-test\np-value = {p_value:.3f} (t-stat = {t_stat:.1f}, df={df})")
plt.xlabel("Giá trị t")
plt.ylabel("Mật độ xác suất")
plt.legend()
plt.grid()
plt.show()
```
